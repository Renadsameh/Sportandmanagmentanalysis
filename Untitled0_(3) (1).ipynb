```python
df['how many hours per week do you spend participating in sports activities? [Insert number of hours]'] = pd.to_numeric(df['how many hours per week do you spend participating in sports activities? [Insert number of hours]'], errors='coerce')
df['Do you believe participating in sports activities helps you manage your time better?  (1 = Poor, 5 = Excellent)'] = pd.to_numeric(df['Do you believe participating in sports activities helps you manage your time better?  (1 = Poor, 5 = Excellent)'], errors='coerce')

df.dropna(subset=['how many hours per week do you spend participating in sports activities? [Insert number of hours]', 'Do you believe participating in sports activities helps you manage your time better?  (1 = Poor, 5 = Excellent)'], inplace=True)

correlation = df['how many hours per week do you spend participating in sports activities? [Insert number of hours]'].corr(df['Do you believe participating in sports activities helps you manage your time better?  (1 = Poor, 5 = Excellent)'])

print("Correlation coefficient:", correlation)

plt.figure(figsize=(8, 6))
plt.scatter(df['how many hours per week do you spend participating in sports activities? [Insert number of hours]'], df['Do you believe participating in sports activities helps you manage your time better?  (1 = Poor, 5 = Excellent)'], color='pink')
plt.xlabel('Hours Spent on Sports')
plt.ylabel('Time Management Rating')
plt.title('Time Management Rating vs. Hours Spent on Sports')
plt.show()
```

```python
Hours_column = df['how many hours per week do you spend participating in sports activities? [Insert number of hours]']
Rate_column = df['On a scale of 1 to 5, how would you rate your ability to manage your time effectively? (1 = Poor, 5 = Excellent)']

# Calculate the mean
mean_Hours = Hours_column .mean()
mean_Rate = Rate_column.mean()
print("Mean of Hours: ", mean_Hours)
print("Mean of Rate: ", mean_Rate)

# Calculate the median
median_Hours = Hours_column .median()
median_Rate = Rate_column.median()
print("Median of Hours:", median_Hours)
print("Median of Rate:", median_Rate)

# Calculate the mode
mode_Hours = Hours_column .mode()[0]
mode_Rate = Rate_column.mode()[0]
print("Mode of Hours:", mode_Hours)
print("Mode of Rate:", mode_Rate)

```
